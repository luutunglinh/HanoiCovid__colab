```python
centroids = np.vstack([df.geometry.centroid.x, df.geometry.centroid.y]).T
```

```python
centroids.shape
```

```python
from matplotlib.lines import Line2D
```

```python
f, ax = plt.subplots(1, figsize=(20, 20))


custom_lines = [Line2D([0], [0], color='#76A3D3', lw=6),
                Line2D([0], [0], color='#FEF391', lw=6),
                Line2D([0], [0], color='#EEB487', lw=6),
                Line2D([0], [0], color='#E87C83', lw=6)]


df.plot(color=df['color'], linewidth=1, ax=ax,edgecolor='gray')

ax.legend(custom_lines, ['Bình thường : 0', 'Nguy cơ : 1-5', 'Nguy cơ cao : 6-20', 'Nguy cơ rất cao : >20'])  
plt.title('BẢN ĐỒ THÔNG TIN DỊCH TỄ COVID-19 THÀNH PHỐ HÀ NỘI')
ax.set_axis_off()

county_centroids_df = df[['geometry']].copy()
county_centroids_df.geometry = df.centroid
county_centroids_df.plot(ax=ax, marker='o', color='lime', markersize=10)


for i in range(0,len(centroids)):
  
  plt.text(centroids[i][0],centroids[i][1],s=f"{df['D_EName'][i]}-{df['patient'][i]}",color='black', fontsize=10,fontweight='bold')

outfp = 'covid19_map.png'
plt.savefig(outfp, dpi=400)

# plt.show()
```

```python
from folium.features import DivIcon
m = folium.Map(location=[21.036898557579672, 105.83179674468283], zoom_start=10, control_scale=True)
for _, r in df.iterrows():
    
    # print(f"{_}-{df['D_EName'][_]}-{df['patient'][_]}")
    a=f"{df['D_EName'][_]}-{df['patient'][_]}"
    b=df['day_14'][_]
    lat = r['centroid'].y
    lon = r['centroid'].x
    # print(f"{lat}-{lon}")
    folium.Marker(location=[ lat,lon],popup = f'covid cộng đồng : {b} ca',icon=DivIcon(
        icon_size=(250,36),
        icon_anchor=(0,0),
        html = """
            <div  style="font-size: 8pt">{}</div>""".format(a)+"""
            
            """
        )).add_to(m)

# m

loc = 'BẢN ĐỒ THÔNG TIN DỊCH TỄ COVID-19 THÀNH PHỐ HÀ NỘI'
title_html = '''
             <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(loc)  
m.get_root().html.add_child(folium.Element(title_html))

#m

from branca.element import Template, MacroElement

template = """
{% macro html(this, kwargs) %}

<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8">
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>jQuery UI Draggable - Default functionality</title>
  <link rel="stylesheet" href="//code.jquery.com/ui/1.12.1/themes/base/jquery-ui.css">

  <script src="https://code.jquery.com/jquery-1.12.4.js"></script>
  <script src="https://code.jquery.com/ui/1.12.1/jquery-ui.js"></script>
  
  <script>
  $( function() {
    $( "#maplegend" ).draggable({
                    start: function (event, ui) {
                        $(this).css({
                            right: "auto",
                            top: "auto",
                            bottom: "auto"
                        });
                    }
                });
});

  </script>
</head>
<body>

 
<div id='maplegend' class='maplegend' 
    style='position: absolute; z-index:9999; border:2px solid grey; background-color:rgba(255, 255, 255, 0.8);
     border-radius:6px; padding: 10px; font-size:14px; right: 20px; bottom: 20px;'>
     
<div class='legend-title'>Chú giải</div>
<div class='legend-scale'>
  <ul class='legend-labels'>
    <li><span style='background:#76A3D3;opacity:0.7;'></span>Bình thường : 0</li>
    <li><span style='background:#FEF391;opacity:0.7;'></span>Nguy cơ : 1-5</li>
    <li><span style='background:#EEB487;opacity:0.7;'></span>Nguy cơ cao : 6-20</li>
     <li><span style='background:#E87C83;opacity:0.7;'></span>Nguy cơ rất cao : >20</li>

  </ul>
</div>
</div>
 
</body>
</html>

<style type='text/css'>
  .maplegend .legend-title {
    text-align: center;
    margin-bottom: 5px;
    font-weight: bold;
    font-size: 90%;
    }
  .maplegend .legend-scale ul {
    margin: 0;
    margin-bottom: 5px;
    padding: 0;
    float: left;
    list-style: none;
    }
  .maplegend .legend-scale ul li {
    font-size: 80%;
    list-style: none;
    margin-left: 0;
    line-height: 18px;
    margin-bottom: 2px;
    }
  .maplegend ul.legend-labels li span {
    display: block;
    float: left;
    height: 16px;
    width: 30px;
    margin-right: 5px;
    margin-left: 0;
    border: 1px solid #999;
    }
  .maplegend .legend-source {
    font-size: 80%;
    color: #777;
    clear: both;
    }
  .maplegend a {
    color: #777;
    }
</style>
{% endmacro %}"""

#m
macro = MacroElement()
macro._template = Template(template)

m.get_root().add_child(macro)


# m.save('map-with-title.html')
folium.GeoJson(df[["geometry","day_14","color"]].to_json(),
    style_function=lambda x: {'fillColor':x["properties"]["color"],'fillOpacity': 0.9, 'opacity': 0.7},
    highlight_function=lambda x: {"fillOpacity": 0.7, "opacity": 0.7}).add_to(m)
m.save('map_covid.html')
m
```
